### The Prediction Problem

Consideriamo coppie di dati $(x,y)$ tali che

$$
    y_i= f(x_i) + \eta_i,
$$

con $f(x_i)$ funzione fissa ma ignota e $\eta_i$ una Gaussiana rappresentante il rumore nel data set.
Vogliamo utilizzare una Neural Network (NN) con dati pesi, parametri e bias per modellizzare i dati in ingresso e provare a fare predizioni.
Sarà dunque necessario fornire alla NN un **training data set** dal cui fit saremo in grado di trovare i migliori parametri, per poi testare ciò che essa ha imparato su un nuovo set di dati, il **validation data set**. 
Per dare una misura della capacità della NN di fare predizioni accurate sui dati, possiamo comparare le predizioni fatte sul data set di test $\{y_j^\mathrm{pred}\}$ con i valori veri $\{y_j\}$, tramite la misura di una Loss Function sul data set di test. Sceglieremo un Optimizer (Optimization Algorithm) che calcoli il gradiente della funzione costo (e quindi la direzione in cui muovere la ricerca) e che sia quindi in grado di trovare i parametri migliori per minimizzarla: più essa è ridotta, meglio la NN è in grado di fare predizioni.


## Esercizio 11.1

Partiamo da un **modello lineare** dato da $$
f(x)=2x+1
$$ 
In tal caso, per il nostro modello basta un solo neurone: ottimizzando il peso dell'unico input ($w$) e il bias ($b$), l'output dipende solo da $wx+b$, ovvero proprio la retta da stimare.
Scegliamo come Loss Function il  Mean Square-Error (MSE): 
$$
MSE= \frac{1}{N_\mathrm{test}}\sum_{j=1}^{N_\mathrm{test}} (y_j^\mathrm{pred}-y_j)^2
$$
e come Optimizer lo Stochastic Gradient Descent (SGD), più in particolare il "mini-batch SGD": fissiamo infatti una batch size pari a $32$, ossia l'algoritmo compara le predizioni con gli output attesi e calcola l'errore alla fine di ciascun batch, e non di ogni sample di dati ($1 < Batch size < N_{train}$).

Vediamo come la regressione NN si modifica variando il numero di epoche $N_{\mathrm{epochs}}\in\{20,50, 100\}$, il numero di dati di training $N_{\mathrm{train}}\in\{100, 500\}$ e il rumore $\sigma\in\{0, 0.5, 1\}$. Ho fissato $N_\mathrm{test}=N_\mathrm{train}/10$.


```python
#target parameters of f(x) = m*x + b we are going to predict
m=2 #slope
b=1 #intersect
np.random.seed(0)
 
#trying different combinations
Epoch=[20, 50, 100] #number of epochs
Train=[100, 500]    #number of training points
Sigma=[0, 0.5, 1.]  #noise

i=0

for sigma in Sigma:
    for epoch in Epoch:
        for train in Train:
            
            #generating training points
            x_train = np.random.uniform(-1, 1, train)
            y_train = np.random.normal(m * x_train + b, sigma)
            #generating validation points
            x_valid = np.random.uniform(-1, 1, int(train/10) )
            x_valid.sort()
            y_valid = np.random.normal(m * x_valid + b, sigma)
            
            #composing the NN model
            model = tf.keras.Sequential()
            model.add(Dense(1, input_shape=(1,)))
            #compiling the NN model
            model.compile(optimizer='sgd', loss='mse', metrics=['acc'])
            #fitting the NN model using training data
            history = model.fit(x=x_train, y=y_train, batch_size=32, epochs=epoch, shuffle=True, validation_data=(x_valid, y_valid), verbose=0)
            
            #preparing predicted points
            x_predicted = np.random.uniform(-1, 1, 100)
            y_predicted = model.predict(x_predicted)
            #and the target function
            y_target = m * x_valid + b
            
           
            #plotting results
            plt.figure(i, figsize = (18, 5))
            #predicted points and the target function
            plt.subplot(1, 2, 1)
            plt.scatter(x_predicted, y_predicted, color='r', label="Predicted")
            plt.plot(x_valid, y_target, label="Target")
            plt.title(f"Target function and predicted points \n N epochs = {epoch} | N train = {train} | $\sigma$ = {sigma}")
            plt.xlabel('x')
            plt.ylabel('y')
            plt.grid(True)
            plt.legend()
            #loss functions
            plt.subplot(1, 2, 2)
            plt.plot(history.history['loss'])
            plt.plot(history.history['val_loss'])
            plt.title('Model loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend(['Train', 'Test'], loc='best')
            plt.grid(True)
            plt.show()
            
            print("Final training data loss: ", history.history['loss'][-1])
            print("Final validation data loss: ", history.history['val_loss'][-1])
            
            # return weights
            print(model.get_weights(), "\n")
            
            i+=1
            print(i, " of ", len(Epoch)*len(Train)*len(Sigma), " done")
            
            
```

Nei casi migliori, i coefficienti che caratterizzano il neurone (peso e bias) sono proprio quelli della retta target. 
Solitamente, i valori più vicini a quelli desiderati si ottengono da una combinazione equilibrata di $N_{train}$ e $N_{epochs}$, non per forza da un aumento squilibrato di uno dei due. Questo ha senso poichè, laddove esiste rumore singificativo, si potrebbe verificare il fenomeno di Overfitting: infatti, nel caso $\sigma=1$ abbiamo il miglior risultato con $N_{epochs}=50$ e $N_{train}=500$ (grafico $16/18$).

In assenza di rumore o con rumore basso invece, si ottiene il migliore fit dei parametri con il maggior numero di dati di training $N_{train}=500$ e di epoche $N_{epochs}=100$ (rispettivamente grafici $6/18$ e $12/18$).
Notiamo che nel caso di $N_{train}$ maggiore, vediamo rispetto al grafico corrispondente ma con un $N_{train}$ minore, una diminuizione dell'errore sul test e un aumento dell'errore sul train, come atteso.

## Esercizio 11.2

Estendiamo ora il modello per ottenere un fit del **polinomio** di terzo ordine:

$$
f(x)=4-3x-2x^2+3x^3
$$
per $x \in [-1,1]$.

Cerchiamo di capire dove entra in gioco la complessità del modello (il suo numero di parametri) e di effettuare una scelta ragionata di:

- numero di layers
- numero di neuroni per layer
- funzione di attivazione
- ottimizzatore
- loss function
  
Ho pensato fosse utile a tal fine confrontare NNs costituite da un diverso numero di Layers $N_{\mathrm{Layers}}\in\{5,10,15\}$, differente numero di neuroni per layer $N_{\mathrm{NeuronsPerLayer}}\in\{5,10,15\}$ ed infine variare la funzione di attivazione, mettendo a confronto una "saturating function" come la $tanh$ con una "non-saturating function" come la $SELU$ (Scaled Exponential Linear Unit). Ho poi scelto $N_{epochs}=50$, $\sigma=0.5$, SGD come Optimizer e MSE come Loss Function. Di nuovo ho fissato $N_\mathrm{test}=N_\mathrm{train}/10$.

```python
#trying different combinations
Layer=[5, 10, 15]       #number of layers
Neuron=[5, 10, 15]      #number of neurons per layer
Active=["selu", "tanh"]  #activation function

epoch=50
sigma=0.5
train=1500

#generating training points
x_train = np.random.uniform(-1, 1, train)
y_train = np.random.normal( 3*x_train**3-2*x_train**2-3*x_train+4, sigma)
#generating validation points
x_valid = np.random.uniform(-1, 1, int(train/10))
x_valid.sort()
y_valid = np.random.normal( 3*x_valid**3-2*x_valid**2-3*x_valid+4, sigma)

i=0

for layer in Layer:
    for neuron in Neuron:
        for active in Active:
            
            #composing the NN model
            model=tf.keras.Sequential()
            model.add(Dense(neuron, input_shape=(1,), activation=active))
            #adding layers
            for k in range(layer-2):
                model.add(Dense(neuron, activation=active))
            model.add(Dense(1))
            #compiling the NN model
            model.compile(optimizer='sgd', loss='mse', metrics=['acc'])
            #fitting the NN model using training data
            history=model.fit(x=x_train, y=y_train, batch_size=32, epochs=epoch, shuffle=True, validation_data=(x_valid, y_valid), verbose=0)
            
            #preparing predicted points
            x_predicted = np.random.uniform(-2, 2, 200)
            y_predicted = model.predict(x_predicted)
            #and the target function
            x_target = np.linspace(-2,2,num=200)
            y_target = 3*x_target**3-2*x_target**2-3*x_target+4
            
            #repeating for the "zoomed in" plot
            x_predicted_short = []
            y_predicted_short = []
            for c in range (0,200) :
                if x_predicted[c]>-1 and x_predicted[c]<1 :
                    x_predicted_short.append(x_predicted[c])
                    y_predicted_short.append(y_predicted[c])
            x_target_short = np.linspace(-1,1,num=100)
            y_target_short = 3*x_target_short**3-2*x_target_short**2-3*x_target_short+4
            
            #plotting results
            plt.figure(i, figsize = (18, 5))
            #predicted points and the target function
            plt.subplot(1, 3, 1)
            plt.scatter(x_predicted, y_predicted, label="Predicted")
            plt.plot(x_target, y_target, color="red", label="Target")
            plt.title(f"Target function and predicted points \n Layers = {layer} | Neurons = {neuron} | Activation = {active}")
            plt.xlabel('x')
            plt.ylabel('y')
            plt.grid(True)
            plt.legend()
            #predicted points and the target function, "zoomed in" version
            plt.subplot(1, 3, 2)
            plt.scatter(x_predicted_short, y_predicted_short, label="Predicted")
            plt.plot(x_target_short, y_target_short, color="red", label="Target")
            plt.title(f"Target function and predicted points \n Layers = {layer} | Neurons = {neuron} | Activation = {active}")
            plt.xlabel('x')
            plt.ylabel('y')
            plt.grid(True)
            plt.legend()
            #loss functions
            plt.subplot(1, 3, 3)
            plt.plot(history.history['loss'])
            plt.plot(history.history['val_loss'])
            plt.title('Model loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend(['Train', 'Test'], loc='best')
            plt.grid(True)
            plt.show()
            
            print("Final training data loss: ", history.history['loss'][-1])
            print("Final validation data loss: ", history.history['val_loss'][-1])
            
            
            i+=1   
            print(i, " of ", len(Layer)*len(Neuron)*len(Active), " done")
```

Alcune osservazioni: innanzitutto, per ambedue la funzioni di attivazione considerate e per tutti i range di $N_{Layers}$ e di $N_{NeuronsPerLayer}$ si ha che tutte le Neural Networks sono in grado di predire correttamente l'andamento dei valori, che risultano molto in accordo con quelli target, all'interno dell'intervallo di Training $[-1,1]$. All'infuori di esso, invece, nessuna delle NN risulta in grado di fare previsioni adeguate.

Per quanto riguarda dunque le previsioni dei dati in generale, non riscontriamo enormi differenze tra le due funzioni di attivazione all'interno dei range considerati per l'analisi.


In linea generale, le NN lavorano in modo ottimale quando la rete è dotata di sufficiente larghezza ma anche di profondità; sono quindi preferibili configurazioni equilibrate, piuttosto che dotate di moltissimi layer da pochi neuroni o viceversa. Viene richiesta inoltre, per il corretto funzionamento della NN, una adeguata quantità di dati su cui lavorare, per evitare i fenomeni di Underfitting e Overfitting: una rete più complessa performerà di sicuro meglio di una rete semplice, ma solo se avrà a sua disposizione un data set di training molto grande. Conviene, pertanto, scegliere per la NN un livello di complessità (e dunque un numero di parametri) adeguato alla quantità di dati di Training a disposizione.

## Esercizio 11.3

Estendiamo ora il nostro modello per fittare la funzione trigonometrica in due variabili $$f(x,y) = \sin(x^2+y^2)$$ nel range $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.

Sulla base di quanto sperimentato finora, deciso di provare ad implementare la NN con:
- $N_{epoch}=100$
- $N_{train}=4000$
- $N_{valid}=400$
- $N_{Layers}=8$
- $N_{NeuronsPerLayer}=14$
- $\sigma=0.2$
- funzione di attivazione: $SELU$
- optimizer: SGD (mini-batch da $Batch Size=32$)

```python
from mpl_toolkits import mplot3d

epoch=100
train=4000
valid=400
layer=8
neuron=14
sigma=0.2
activate="selu"

#generating training points in the right range of values
x_train=np.random.uniform(-1.5, 1.5, train)
y_train=np.random.uniform(-1.5, 1.5, train)
z_train=np.random.normal(np.sin(x_train**2+y_train**2), sigma)
#generating validation points in the right range of values
x_valid=np.random.uniform(-1.5, 1.5, valid)
y_valid=np.random.uniform(-1.5, 1.5, valid)
x_valid.sort()
y_valid.sort()
z_valid=np.random.normal(np.sin(x_valid**2+y_valid**2), sigma)

#composing the NN model
model=tf.keras.Sequential()
model.add(Dense(neuron, input_shape=(2,), activation=activate))
#adding layers
for k in range(layer-2):
        model.add(Dense(neuron, activation=activate))
model.add(Dense(1))
#compiling the NN model
model.compile(optimizer='sgd', loss='mse', metrics=['acc'])
#fitting the NN model using training data
history=model.fit(x=np.array([x_train, y_train]).T, y=z_train, batch_size=32, epochs=epoch, shuffle=True, validation_data=(np.array([x_valid, y_valid]).T, z_valid), verbose=0)

#preparing predicted points
x_predicted=np.random.uniform(-1.5, 1.5, 2000)
y_predicted=np.random.uniform(-1.5, 1.5, 2000)
z_predicted=model.predict(np.array([x_predicted, y_predicted]).T)
#and the target function
x=np.outer(np.linspace(-1.5,1.5,30), np.ones(30))
y=x.copy().T
z=np.sin(x**2+y**2)

#plotting results
#loss functions
plt.figure(figsize = (13, 9))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(True)

print("Final training data loss: ", history.history['loss'][-1])
print("Final validation data loss: ", history.history['val_loss'][-1])

#predicted points and the target function
plt.figure(figsize = (15, 10))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='hsv', alpha=0.2, label='Target')
ax.scatter(x_predicted, y_predicted, z_predicted, color='purple', label='Predicted')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Target surface and predicted points")
ax.view_init(45,45)

plt.show()
```

Entrambi gli errori calano dopo un numero ragionevole di epoche. Nella figura 3D vediamo raffigurata la superficie target, in viola i punti predetti con la NN: sembrano giacere in buona approssimazione sulla superficie target.
