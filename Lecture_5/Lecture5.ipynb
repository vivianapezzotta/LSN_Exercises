Ci proponiamo di stimare la distanza media dall'origine di un elettrone all'interno di un atomo di idrogeno, prima nel Ground State 1s rappresentato dalla funzione d'onda $\Psi_{1,0,0}$ e poi nello stato eccitato 2p rappresentato invece da $\Psi_{2,1,0}$. Sappiamo inoltre che:
$$
\psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
e
$$
\psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$

Ad esse associamo le corrispettive distribuzioni di probabilità in coordinate cartesiane date da
$$p_{1,0,0}(x,y,z) = |{\psi}_{1,0,0}(x,y,z)|^2 \propto e^{-\frac{2\sqrt{x^2+y^2+z^2}}{a_0}}$$
e $$p_{2,1,0}(x,y,z) = |{\psi}_{2,1,0}(x,y,z)|^2 \propto (x^2+y^2+z^2)e^{-\frac{\sqrt{x^2+y^2+z^2}}{a_0}}\frac{z^2}{x^2+y^2+z^2}=z^2e^{-\frac{\sqrt{x^2+y^2+z^2}}{a_0}}$$

Utilizziamo l'algoritmo di Metropolis per campionare le due distribuzioni di probabilità, nei due casi di
- probabilità di transizione $T(\vec{x}|\vec{y})$ uniforme
- probabilità di transizione $T(\vec{x}|\vec{y})$ gaussiana

Le posizioni saranno utilizzate per stimare $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ e $\left\langle r \right\rangle_{\Psi_{2,1,0}}$, da confrontare poi con i valori teorici attesi:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$

Per trovare i valori dello step della probabilità di transizione uniforme e gaussiana $\delta$, tali che il rate di accettazione definito come $A=\min \left(1,\frac{p(\vec{x})}{p(\vec{y})}\right)$ risulti attorno al $50\%$, si eseguono alcune simulazioni prova con un numero elevato di punti (nel nostro caso $M=1000$) modificando il valore del parametro fino ad ottenere il risultato desiderato.
Ho ritenuto adeguati tali valori:
- distribuzione uniforme : $\delta_{1,0,0}=1.27$ e $\delta_{2,1,0}=0.76$
- distribuzione gaussiana: $\delta_{1,0,0}=2.95$ e $\delta_{2,1,0}=1.89$

Procediamo ora ad equilibrare il nostro sistema, in modo tale che l'elettrone si trovi già in una regione con alta densità di probabilità. Per fare ciò, eseguiamo una simulazione con $M=1000$ punti: partendo da un punto lontano dalle zone ad alta probabilità delle due distribuzioni, si nota che nelle prime fasi del campionamento vediamo una scia di punti diretta verso la zona a probabilità maggiore. Di seguito i grafici ottenuti nel caso uniforme, dove ho impostato come punto di partenza le coordinate cartesiane $(25,25,25)$:


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X,Y,Z=np.loadtxt( "EQ/EQpoints100Unif.out", usecols=(0,1,2), delimiter=' ', unpack='true' )
X2,Y2,Z2=np.loadtxt( "EQ/EQpoints210Unif.out", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig = plt.figure(figsize=(22,10))

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x [$a_0$]')
ax.set_ylabel('y [$a_0$]')
ax.set_zlabel('z [$a_0$]')
ax.title.set_text("$ \psi_{1,0,0}$ Uniform")
ax.view_init(0, 0)
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(X2, Y2, Z2, c=Z2, marker='.')
ax.set_xlabel('x [$a_0$]')
ax.set_ylabel('y [$a_0$]')
ax.set_zlabel('z [$a_0$]')
ax.title.set_text("$\psi_{2,1,0}$ Uniform")
ax.view_init(0, 0)
plt.show()
```

Come vediamo anche dalle scie evidenti nei grafici, un buon punto di partenza nel caso di $ \psi_{1,0,0}$ (ground state 1s) è l'origine,  mentre nel caso di $\psi_{2,1,0}$ (stato eccitato 2p) l'origine ha probabilità nulla e consideriamo ottimale un punto come $(0,0,2)$.

```python
x, y = np.loadtxt( "EQ/EQvalues100Unif.out", usecols=(0,1), delimiter=' ', unpack='true' )
x2, y2 = np.loadtxt( "EQ/EQvalues210Unif.out", usecols=(0,1), delimiter=' ', unpack='true' )

plt.figure(figsize=(15,7))
plt.hlines(1.5,0,1E3, color='orange', label='1s Expected', linestyle="dashed")
plt.hlines(5,0,1E3, color='green', label='2p Expected', linestyle="dashed")
plt.errorbar(x,y,label='1s', color='orange')
plt.errorbar(x2,y2,label='2p', color='green')
plt.title('Equilibration Values')
plt.xlabel('Number of Steps')
plt.ylabel("r [$a_0$]")
plt.legend()
plt.grid(True)

plt.show()
```

Dal grafico notiamo che nel caso della funzione d'onda $\psi_{2,1,0}$ già dopo circa 100 step raggiungiamo la distanza media di equilibrio, mentre nel caso di $\psi_{1,0,0}$ si arriva all'equilibrio dopo circa 250 step.

Procediamo ora definendo da quanti step dovrebbe essere formato ogni blocco. Plottiamo l'autocorrelazione calcolata sui valori istantanei della distanza in un campione di 100000 punti, nel caso uniforme di $\psi_{1,0,0}$:

```python
r = np.loadtxt('EQ/AutocorrValues100Unif.out', usecols=(1))
autocorr = []
for tau in range(100):
    l = len(r) - tau
    r_t = 0
    r_tau = 0
    r_ttau = 0
    r_t2 = 0
    for t in range(l):
        r_t += r[t]
        r_tau += r[t+tau]
        r_ttau += r[t]*r[t+tau]
        r_t2 += r[t]**2
    r_t /= l
    r_tau /= l
    r_ttau /= l
    r_t2 /= l
    autocorr.append((r_ttau-r_t*r_tau)/(r_t2-r_t**2))

plt.figure(figsize = (15,5))    
plt.plot(range(len(autocorr)),autocorr)
plt.title('Autocorrelation')
plt.xlabel('Time Step t')
plt.ylabel("Autocorrelation")
plt.grid()
plt.show()
```

Vediamo che il valore della funzione di autocorrelazione è pressochè nullo già con un Time Step di $60$. Per maggiore sicurezza, possiamo quindi considerare blocchi da $100$.
