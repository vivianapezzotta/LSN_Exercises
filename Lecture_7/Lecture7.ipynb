## Autocorrelazione

Vogliamo ora avere un'idea della larghezza di un blocco da usare nelle simulazioni (ossia del numero di step che lo compongono). Ciò deriva dalla necessità, al fine di rendere il metodo del Data Blocking efficiente, di dividere le misure in blocchi che hanno valori medi non correlati, per evitare così una sottostima dell'incertezza.

Plottiamo quindi l'autocorrelazione dei dati di energia potenziale e pressione nelle tre fasi: solida, liquida e gassosa.

```python
def AutoCorrelation(x, tau):
    
    l = len(x)-tau
    x_t = 0
    x_tau = 0
    x_ttau = 0
    x_t2 = 0
    
    for t in range(l):
        x_t  += x[t]
        x_tau += x[t+tau]
        x_ttau += x[t]*x[t+tau]
        x_t2  += x[t]**2
        
    x_t /= l
    x_tau /= l
    x_ttau /= l
    x_t2 /= l
    
    return (x_ttau-x_t*x_tau)/(x_t2-x_t**2)
```

```python
Usolid=np.loadtxt("INSTANT_VAL/SOLID_output.epot.0")
Psolid=np.loadtxt("INSTANT_VAL/SOLID_output.pres.0")
Uliquid=np.loadtxt("INSTANT_VAL/LIQUID_output.epot.0")
Pliquid=np.loadtxt("INSTANT_VAL/LIQUID_output.pres.0")
Ugas=np.loadtxt("INSTANT_VAL/GAS_output.epot.0")
Pgas=np.loadtxt("INSTANT_VAL/GAS_output.pres.0")

CorrUsolid = []
CorrPsolid = []
CorrUliquid = []
CorrPliquid = []
CorrUgas = []
CorrPgas = []
```

```python
print("This will take a while...")
for tau in range(400):
    
    CorrUsolid.append(AutoCorrelation(Usolid,tau))          #Internal energy autocorrelation
    CorrUliquid.append(AutoCorrelation(Uliquid,tau)) 
    CorrUgas.append(AutoCorrelation(Ugas,tau))
    
    CorrPsolid.append(AutoCorrelation(Psolid,tau))          #Pressure autocorrelation
    CorrPliquid.append(AutoCorrelation(Pliquid,tau))
    CorrPgas.append(AutoCorrelation(Pgas,tau))

tau = np.arange(400)
```

```python
plt.figure(1, figsize = (13, 6))
plt.title("Autocorrelation Solid Phase")
plt.xlabel(r'$\tau$')
plt.ylabel('$Autocorrelation$')
plt.plot(tau,CorrUsolid,label="$U/N$")
plt.plot(tau,CorrPsolid,label="$Pressure$")
plt.grid(True)
plt.legend()
```

Visualizziamo ora come cambia il valore dell'incertezza statistica calcolata secondo il metodo del Data Blocking col variare degli step che costituiscono un singolo blocco: $L \in \left[ 10; 5000 \right]$.

```python
def blocking_method_error(x, L):
    
    M = len(x) #tot data
    N = int(M/L)  #number of blocks
    
    ave = np.zeros(N)
    ave2=np.zeros(N)
    
    for i in range(N): #cycle on the blocks
        
        blk_ave=0

        for j in range(L):  #cycle inside every block
            blk_ave += x[j+L*i]
            
        ave[i] = blk_ave/L  #block average
        ave2[i]=(ave[i])**2
        
    accu = 0
    accu2 = 0

    for i in range(N):
        accu += ave[i]  #cumulative average
        accu2+= ave2[i]
    
    accu /= N  #final value
    accu2 /= N
        
    return np.sqrt(abs(accu2-(accu**2))/N)
```

```python
errU = []
errp = []

for blk_size in range(10, 5000, 50):
    errU.append(blocking_method_error(Usolid, blk_size))
    errp.append(blocking_method_error(Psolid, blk_size))
print('\t','\t','\t','\t','\t','\t','   ', "SOLID PHASE")
    
plt.subplot(1,2,1)
plt.plot(range(10, 5000, 50), errU, label="$U/N$", color='blue')
plt.title('Energy Uncertainty')
plt.xlabel('Block Size')
plt.ylabel('Statistical Uncertainty')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(range(10, 5000, 50), errp, label="Pressure", color='turquoise')
plt.title('Pressure Uncertainty')
plt.xlabel('Block Size')
plt.ylabel('Statistical Uncertainty')
plt.grid(True)

plt.subplots_adjust(left=0.2, bottom=0.2, right=2.2, top=1.2, wspace=0.3, hspace=0.4) 

plt.show()
```

Possiamo osservare, in modo simile nelle tre fasi e per entrambe le quantità in esame, una crescita iniziale nel valore delle incertezze: ciò è motivato dal fatto che per dimensioni del blocco troppo piccole i valori considerati risultano correlati tra loro e si ha dunque una sottostima dell'incertezza. Inoltre, dopo un certo numero di step nel blocco (circa $2000$) avviene una stabilizzazione del valore dell'incertezza, che rimane pressochè costante all'aumentare della lunghezza del blocco (i dati sono praticamente del tutto scorrelati tra loro).

```python
errUSOLID = []
errpSOLID = []
errULIQUID = []
errpLIQUID = []
errUGAS = []
errpGAS = []

for blk_size in range(10, 500, 50):
    errUSOLID.append(blocking_method_error(Usolid, blk_size))
    errpSOLID.append(blocking_method_error(Psolid, blk_size))
    errULIQUID.append(blocking_method_error(Uliquid, blk_size))
    errpLIQUID.append(blocking_method_error(Pliquid, blk_size))
    errUGAS.append(blocking_method_error(Ugas, blk_size))
    errpGAS.append(blocking_method_error(Pgas, blk_size))
    

    
plt.subplot(3,2,1)
plt.plot(range(10, 500, 50), errUSOLID, label="$U/N$", color='blue')
plt.title('Energy Uncertainty')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid(True)

plt.subplot(3,2,2)
plt.plot(range(10, 500, 50), errpSOLID, label="Pressure", color='turquoise')
plt.title('Pressure Uncertainty')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid(True)

plt.subplot(3,2,3)
plt.plot(range(10, 500, 50), errULIQUID, label="$U/N$", color='darkgreen')
plt.title(' ')
plt.xlabel(' ')
plt.ylabel('Statistical Uncertainty')
plt.grid(True)

plt.subplot(3,2,4)
plt.plot(range(10, 500, 50), errpLIQUID, label="Pressure", color='lightgreen')
plt.title(' ')
plt.xlabel(' ')
plt.ylabel('Statistical Uncertainty')
plt.grid(True)

plt.subplot(3,2,5)
plt.plot(range(10, 500, 50), errUGAS, label="$U/N$", color='red')
plt.title(' ')
plt.xlabel('Block Size')
plt.ylabel(' ')
plt.grid(True)

plt.subplot(3,2,6)
plt.plot(range(10, 500, 50), errpGAS, label="Pressure", color='orange')
plt.title(' ')
plt.xlabel('Block Size')
plt.ylabel('')
plt.grid(True)

plt.subplots_adjust(left=0.2, bottom=0.2, right=2.2, top=1.2, wspace=0.3, hspace=0.4) 

plt.show()
```

## Funzione di Distribuzione Radiale $g(r)$

La funzione di distribuzione radiale rappresenta la probabilità di trovare una coppia di particelle a distanza $r$, relativa alla probabilità attesa per una distribuzione totalmente random alla stessa densità. Per calcolare $g(r)$ data una certa configurazione si "contano" le particelle che si trovano a una distanza compresa tra $r$ e $r+dr$ realizzando un istogramma, che viene poi normalizzato.
Nel nostro caso, consideriano $nbins =100$.

Abbiamo implementato il calcolo della funzione di distribuzione radiale sia nel codice per l'insieme canonico $NVT MC$ sia in quello per l'insieme microcanonico (Molecular Dynamics) $NVE$. Vogliamo ora confrontare i risultati ottenuti, pur notando che nel primo caso l'algoritmo campiona delle configurazioni da una certa distribuzione di probabilità (determinata dai pesi di Boltzmann), nel secondo caso invece le particelle vengono fatte evolvere secondo le equazioni del moto e analizzate nel tempo. 

```python
r, gofr, err_gofr = np.loadtxt("SIM/NVT_SOLID_output.gave.0", usecols=(1,2,3), unpack=True)
gofr_NVE, err_gofr_NVE = np.loadtxt("NVE/NVE_SOLID_output_g.out", usecols=(2,3), unpack=True)
r*=sigma
plt.figure(1, figsize = (15, 6))
plt.errorbar(r, gofr, err_gofr, label="NVT MC")
plt.errorbar(r, gofr_NVE, err_gofr_NVE, label="NVE")
plt.title("SOLID PHASE")
plt.xlabel("$r[m]$")
plt.ylabel("$g(r)$")
plt.grid(True)
plt.legend()
plt.show()
```

Dai grafici vediamo che la distribuzione della distanza tra le particelle per ciascuna delle tre fasi considerate è analoga utilizzando l'uno o l'altro algoritmo. Pare dunque sensato attendersi che anche microscopicamente i due sistemi considerati siano simili, essendo stati posti in simili condizioni termodinamiche.

Notiamo che, indipendentemente dalla fase considerata, $g(r)$ risulta nulla fino a circa $r=\sigma$: questo accade a causa dell'incompenetrabilità della materia, giustificata dalla parte fortemente repulsiva del potenziale di Lennard-Jones per piccole distanze. Di fatto, la probabilità di trovare due particelle a tali distanze è resa nulla dalla repulsività del potenziale, fino al valore in cui esso cambia di segno, ovvero $\sigma$.

In tutti e tre i casi osserviamo poi un picco, che corrisponde al minimo del potenziale di Lennard-Jones, a circa $r\approx 1.122 \sigma$ (la linea rossa tratteggiata nei grafici). Nella fase liquida e solida il massimo di $g(r)$ si trova in realtà un po' prima rispetto alla distanza per cui si ha il minimo di potenziale, mentre il massimo della fase gassosa sta esattamente a $r\approx1.122\sigma$.

Si può inoltre notare come nel solido emerga la struttura cristallina del sistema: si alternano periodicamente distanze a cui è molto probabile trovare una coppia di particelle a zone in cui invece la probabilità rasenta lo 0. 
Nella fase gassosa, invece, vediamo che all'aumentare della distanza $g(r) \approx 1$, ovvero il gas di Lennard-Jones diventa approssimabile con un gas ideale.
