Ci proponiamo di simulare un sistema ben rappresentato dal modello di Ising 1D, composto da $N=50$ spin unitari interagenti ognuno con i suoi primi vicini, dove il parametro di interazione vale $J=1$ e inoltre $\mu_B=1$ e $k_B=1$. Il range di temperatura dell'insieme canonico all'interno del quale abbiamo scelto di lavorare è $[0.5, 2.0]$. Effettueremo le simulazioni campionando il sistema sia con l'algoritmo di Metropolis sia con quello di Gibbs.

Riporto a seguire le soluzioni analitiche esatte, note grazie al fatto che il modello unidimensionale di Ising è risolto, in modo da agevolare un confronto con i nostri risultati.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

Ns = 50
J=1.0


def u(T):
    th = np.tanh(J/T)
    thN= th**Ns
    ch = 1/th
    e = -J*( th + ch*thN )/( 1 + thN )
    return e

def c(T):
    beta = 1/T
    th = np.tanh(J/T)
    thN= th**Ns
    ch = 1/th
    heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
    return heat

def m(T,h):
    b = 1/T

    l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    Z = l1**Ns + l2**Ns
    M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
            + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
    return M
    
def chi(T):
    th = np.tanh(J/T)
    beta = 1/T
    thN = th**Ns
    chi = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
    return chi
```

Innanzitutto, tramite simulazioni composte da $500$ step andiamo a verificare il numero di step precedenti alla simulazione vera e propria necessari affinchè il sistema raggiunga l'equilibrio, con entrambi gli algoritmi e alle due temperature estreme (per evitare di visualizzare dati relativi al sistema in uno stato metastabile):

```python
xM = np.linspace(1,500,num=500)
yM = np.loadtxt( "EQ/EQ_METRO2_output.ene.0", usecols=(1), unpack='true', max_rows=500 )
xG = np.linspace(1,500,num=500)
yG = np.loadtxt( "EQ/EQ_GIBBS2_output.ene.0", usecols=(1), unpack='true', max_rows=500)


fig, (ax1, ax2) = plt.subplots(2, 1,  figsize=(15,10), dpi=180)
fig.suptitle('EQUILIBRATION for T=2')

ax1.plot(xM, yM, color='green')
ax1.set_xlabel('Number of Steps')
ax1.set_ylabel('U/N')
ax1.set_title('Metropolis')
ax1.grid(True)

ax2.plot(xG, yG, color='blue')
ax2.set_xlabel('Number of Steps')
ax2.set_ylabel('U/N')
ax2.set_title('Gibbs')
ax2.grid(True)

plt.show()

```

Risulta quindi dai grafici che in tutti i casi, senza apprezzabili differenze tra i due algoritmi, il sistema raggiunge l'equilibrio piuttosto in fretta, a meno di 100 step. Per essere certi di aver portato correttamente a termine la fase di equilibrazione, possiamo perciò impostare $150$ step precedenti alla simulazione vera e propria.

Procediamo ora con la simulazione: consideriamo $N=20$ blocchi da $10000$ step ciascuno e analizziamo l'andamento, in funzione della temperatura, dell'energia interna, della capacità termica, della suscettività (a campo esterno nullo) ed infine della magnetizzazione (a campo esterno non nullo ma pari ad $h=0.02$).

Consideriamo la temperatura nel range $[0.5, 2.0]$, a passi di $\Delta T = 0.05$.

Per tutte le grandezze, plottiamo a sinistra i dati ottenuti con l'algoritmo di Metropolis, quelli ottenuti con l'algoritmo di Gibbs e la curva teorica attesa; a destra invece possiamo vedere la discrepanza tra le curve ottenute dalle simulazioni MC per Metropolis e Gibbs rispetto alla curva teorica, con le rispettive incertezze.

```python
points=100
Tmodel = np.linspace(0.2,3.0,num=points)
umodel=u(Tmodel)

Tmetro, umetro, err_umetro=np.loadtxt( "FINAL/METRO_ene.dat", usecols=(0,1,2), delimiter=' ', unpack='true' )
Tgibbs, ugibbs, err_ugibbs=np.loadtxt( "FINAL/GIBBS_ene.dat", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig, ax=plt.subplots(1,2, figsize=(15,7), dpi=120)

ax[0].plot(Tmodel, umodel, label="Analytical model", color="red")
ax[0].errorbar(Tmetro, umetro, yerr=err_umetro, label="Metropolis", color="blue")
ax[0].errorbar(Tgibbs, ugibbs, yerr=err_ugibbs, label="Gibbs", color="green")
ax[0].set_title("Ising 1D - Internal energy")
ax[0].set_xlabel("T")
ax[0].set_ylabel("U/N")
ax[0].grid(True)
ax[0].legend()

ax[1].set_title("Ising 1D - Internal Energy Discrepance (with uncertainties)")
ax[1].set_xlabel("T")
ax[1].set_ylabel("U/N - $U/N_{model}$")
ax[1].grid(True)
ax[1].axhline(0, 0,1, label="Expected", color="red")
ax[1].errorbar(Tmetro, umetro-u(Tmetro),yerr=err_umetro, label="Metropolis", color="blue", capsize=3)
ax[1].errorbar(Tgibbs, ugibbs-u(Tgibbs),yerr=err_ugibbs, label="Gibbs", color="green", capsize=3)
ax[1].legend()
    
plt.show()
```

```python
points=100
Tmodel = np.linspace(0.2,3.0,num=points)
cmodel=c(Tmodel)

Tmetro, cmetro, err_cmetro=np.loadtxt( "FINAL/METRO_heat.dat", usecols=(0,1,2), delimiter=' ', unpack='true' )
Tgibbs, cgibbs, err_cgibbs=np.loadtxt( "FINAL/GIBBS_heat.dat", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig, ax=plt.subplots(1,2, figsize=(15,7), dpi=120)

ax[0].plot(Tmodel, cmodel, label="Analytical model", color="red")
ax[0].errorbar(Tmetro, cmetro, yerr=err_cmetro, label="Metropolis", color="blue")
ax[0].errorbar(Tgibbs, cgibbs, yerr=err_cgibbs, label="Gibbs", color="green")
ax[0].set_title("Ising 1D - Heat capacity")
ax[0].set_xlabel("T")
ax[0].set_ylabel("C/N")
ax[0].grid(True)
ax[0].legend()

ax[1].set_title("Ising 1D - Heat capacity Discrepance (with uncertainties)")
ax[1].set_xlabel("T")
ax[1].set_ylabel("C/N - $C/N_{model}$")
ax[1].grid(True)
ax[1].axhline(0, 0,1, label="Expected", color="red")
ax[1].errorbar(Tmetro, cmetro-c(Tmetro),yerr=err_cmetro, label="Metropolis", color="blue", capsize=3)
ax[1].errorbar(Tgibbs, cgibbs-c(Tgibbs),yerr=err_cgibbs, label="Gibbs", color="green", capsize=3)
ax[1].legend()
    
plt.show()
```

Per calcolare la capacità termica abbiamo misurato $\langle H^2 \rangle$ e calcolato $C$ soltanto alla fine della misura di ogni blocco, in quanto la capacità termica è legata a fluttuazioni quadratiche dell'energia e non è quindi possibile considerare misurazioni istantanee.

```python
points=100
Tmodel = np.linspace(0.2,3.0,num=points)
mmodel=m(Tmodel,0.02)

Tmetro, mmetro, err_mmetro=np.loadtxt( "FINAL/METRO_mag.dat", usecols=(0,1,2), delimiter=' ', unpack='true' )
Tgibbs, mgibbs, err_mgibbs=np.loadtxt( "FINAL/GIBBS_mag.dat", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig, ax=plt.subplots(1,2, figsize=(15,7), dpi=120)

ax[0].plot(Tmodel, mmodel, label="Analytical model", color="red")
ax[0].errorbar(Tmetro, mmetro, yerr=err_mmetro, label="Metropolis", color="blue")
ax[0].errorbar(Tgibbs, mgibbs, yerr=err_mgibbs, label="Gibbs", color="green")
ax[0].set_title("Ising 1D - Magnetization")
ax[0].set_xlabel("T")
ax[0].set_ylabel("M/N")
ax[0].grid(True)
ax[0].legend()

ax[1].set_title("Ising 1D - Magnetization Discrepance (with uncertainties)")
ax[1].set_xlabel("T")
ax[1].set_ylabel("M/N - $M/N_{model}$")
ax[1].grid(True)
ax[1].axhline(0, 0,1, label="Expected", color="red")
ax[1].errorbar(Tmetro, mmetro-m(Tmetro, 0.02),yerr=err_mmetro, label="Metropolis", color="blue", capsize=3)
ax[1].errorbar(Tgibbs, mgibbs-m(Tgibbs, 0.02),yerr=err_mgibbs, label="Gibbs", color="green", capsize=3)
ax[1].legend()
    
plt.show()
```

La magnetizzazione è stata valutata aggiungendo un campo esterno $h=0.02$ nelle simulazioni.

```python
points=100
Tmodel = np.linspace(0.2,3.0,num=points)
chimodel=chi(Tmodel)

Tmetro, chimetro, err_chimetro=np.loadtxt( "FINAL/METRO_chi.dat", usecols=(0,1,2), delimiter=' ', unpack='true' )
Tgibbs, chigibbs, err_chigibbs=np.loadtxt( "FINAL/GIBBS_chi.dat", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig, ax=plt.subplots(1,2, figsize=(15,7), dpi=120)

ax[0].plot(Tmodel, chimodel, label="Analytical model", color="red")
ax[0].errorbar(Tmetro, chimetro, yerr=err_chimetro, label="Metropolis", color="blue")
ax[0].errorbar(Tgibbs, chigibbs, yerr=err_chigibbs, label="Gibbs", color="green")
ax[0].set_title("Ising 1D - Susceptibility")
ax[0].set_xlabel("T")
ax[0].set_ylabel("$\chi/N$")
ax[0].grid(True)
ax[0].legend()

ax[1].set_title("Ising 1D - Susceptibility - uncertainties")
ax[1].set_xlabel("T")
ax[1].set_ylabel("$\chi/N$ - $\chi /N_{model}$")
ax[1].grid(True)
ax[1].axhline(0, 0,1, label="Expected", color="red")
ax[1].errorbar(Tmetro, chimetro-chi(Tmetro),yerr=err_chimetro, label="Metropolis", color="blue", capsize=3)
ax[1].errorbar(Tgibbs, chigibbs-chi(Tgibbs),yerr=err_chigibbs, label="Gibbs", color="green", capsize=3)
ax[1].legend()
    
plt.show()
```

Ritengo opportuno segnalare che, seppure la suscettività magnetica si calcoli come $ \beta\langle (\sum_i s_i)^2\rangle-\langle \sum_{s_i}\rangle^2$, a campo esterno $h=0$ il secondo termine risulta nullo in quanto $\langle \sum_{s_i}\rangle=0$. Dunque abbiamo calcolato la suscettività secondo la formula $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle$.

Possiamo dedurre dai grafici una considerazione generale: i valori ottenuti dalle simulazioni risultano per lo più in accordo con i valori teorici attesi; 
inoltre, le incertezze aumentano a mano a mano che la temperatura diminuisce.
Notiamo che l'algoritmo di Gibbs risulta meno performante del Metropolis a basse temperature.

Infine, esaminiamo il comportamento del parametro di Acceptance Rate al variare della temperatura, nel caso di campionamento con algoritmo di Metropolis (per Gibbs infatti esso è sempre pari a 1) nel caso $h=0$:

```python
x, y = np.loadtxt( "ACC/AccRate.dat", usecols=(0,1), delimiter=' ', unpack='true' )

fig=plt.figure(figsize=(15,5))
plt.plot(x,y, color='purple')
plt.title('Acceptance Rate vs Temperature')
plt.xlabel('T')
plt.ylabel('Acceptance Rate')
plt.grid(True)
plt.show()
```

Al diminuire della temperatura, anche l'Acceptance Rate diminuisce, come previsto: infatti, abbassando la temperatura risulta meno probabile trovare una configurazione degli spin accettabile (andando verso $T=0$ si ottengono le due configurazioni limite di spin tutti allineati up o down).
