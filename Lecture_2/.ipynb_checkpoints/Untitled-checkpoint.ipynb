```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, sum_prog, err_prog = np.loadtxt( "data_unif.out", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig=plt.figure(figsize=(15,5))
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.title('Media')
plt.xlabel('#lanci')
plt.ylabel('<r>')
plt.grid(True)
plt.show()
```
