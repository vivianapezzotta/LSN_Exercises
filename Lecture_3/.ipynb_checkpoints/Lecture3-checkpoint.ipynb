Ci proponiamo di calcolare con il metodo MC il prezzo a t=0 di un'opzione Europea CALL e un'opzione Europea PUT, dati rispettivamente da
$$C[S(0),0]=\langle e^{-rT}\max[0,S(T)-K]\rangle$$
e
$$P[S(0),0]=\langle e^{-rT}\max[0,K-S(T)]\rangle$$
utilizzando i seguenti parametri: 
- asset price al tempo $t=0$: $S(0)=100$
- strike price: $K=100$
- delivery time: $T=1$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

Effettuiamo il calcolo seguendo prima un processo diretto, ovvero campionando l'asset price finale S(T) al delivery time:
 $$S(T)=S(0)\exp\left[\left( r-\frac{1}{2}\sigma^2 \right)T+\sigma W(t)\right], \quad W(t) \sim N(0,t)$$
In seguito discretizziamo il processo, dividendo $[0,T]$ in $10^2$ intervalli temporali e calcolando step by step S(t) in modo ricorsivo da:
 $$S(t_{i+1})=S(t_i)\exp\left[\left( r-\frac{1}{2}\sigma^2 \right)(t_{i+1}-t_{i})+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right], \quad Z_i \sim N(0,1) $$


Eseguiamo $M=10^5$ simulazioni divise in in $N=100$ blocchi da $L=1000$ ognuna.
Possiamo inoltre confrontare i risultati ottenuti via MC con quelli derivati da calcoli analitici secondo Black e Scholes:

```python
import numpy, scipy
from scipy.special import erf
from numpy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

A seguire i risultati ottenuti dalle simulazioni MC nei 4 casi di:

- European Call Option (Direct)
- European Call Option (Discrete)
- European Put Option (Direct)
- European Put Option (Discrete)

```python
import numpy as np
import matplotlib.pyplot as plt

xC_direct,yC_direct, errorC_direct=np.loadtxt("DirectCall.out", usecols=(0,1,2), delimiter=" ", unpack='true')
xC_discrt,yC_discrt, errorC_discrt=np.loadtxt("DiscreteCall.out", usecols=(0,1,2), delimiter=" ", unpack='true')
xP_direct,yP_direct, errorP_direct=np.loadtxt("DirectPut.out", usecols=(0,1,2), delimiter=" ", unpack='true')
xP_discrt,yP_discrt, errorP_discrt=np.loadtxt("DiscretePut.out", usecols=(0,1,2), delimiter=" ", unpack='true')
fig = plt.figure(figsize=(16,16))

ax1 = fig.add_subplot(221)
ax1.hlines(call,0,1E5, color='red',label="Analytic solution")
ax1.errorbar(xC_direct,yC_direct,yerr=errorC_direct, color='lightblue')
ax1.set_title("Direct Call Option price")
ax1.set_xlabel('Number of Throws')
ax1.set_ylabel("C[S(0),0]")
ax1.legend()
ax1.grid(True)

ax2 = fig.add_subplot(222)
ax2.hlines(call,0,1E5, color='red',label="Analytic solution")
ax2.errorbar(xC_discrt,yC_discrt,yerr=errorC_discrt, color='blue')
ax2.set_title('Discrete Call Option price')
ax2.set_xlabel('Number of Throws')
ax2.set_ylabel("C[S(0),0]")
ax2.legend()
ax2.grid(True)

ax3 = fig.add_subplot(223)
ax3.hlines(put,0,1E5, color='red',label="Analytic solution")
ax3.errorbar(xP_direct,yP_direct,yerr=errorP_direct, color='lightgreen')
ax3.set_title('Direct Put Option price')
ax3.set_xlabel('Number of Throws')
ax3.set_ylabel("P[S(0),0]")
ax3.legend()
ax3.grid(True)

ax4 = fig.add_subplot(224)
ax4.hlines(put,0,1E5, color='red',label="Analytic solution")
ax4.errorbar(xP_discrt,yP_discrt,yerr=errorP_discrt, color='darkgreen')
ax4.set_title('Discrete Put Option price')
ax4.set_xlabel('Number of Throws')
ax4.set_ylabel("P[S(0),0]")
ax4.legend()
ax4.grid(True)

plt.show()

print('Last block')
print('Direct Call Option price: ', yC_direct[99], '±', errorC_direct[99])
print('Discrete Call Option price: ', yC_discrt[99], '±', errorC_discrt[99])
print('Direct Put Option price: ', yP_direct[99], '±', errorP_direct[99])
print('Discrete Put Option price: ', yP_discrt[99], '±', errorP_discrt[99])
print(" == BLACK-SCHOLES ==")
print("Call: ", call)
print("Put: ", put)
```

Possiamo vedere dai grafici che il valore finale ottenuto dalle simulazioni MC in entrambi i casi (diretto e discreto) risulta compatibile entro l'incertezza statistica con il valore trovato dal calcolo analitico esatto.
